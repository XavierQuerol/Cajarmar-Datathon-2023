```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_meteo = pd.read_csv("./ds_tractats/df_meteo_tractat.csv", index_col = False)
df_eto = pd.read_csv("./ds_tractats/df_eto_tractat.csv", index_col = False)
```

```python
df_meteo_eto = pd.merge(df_meteo, df_eto, left_on=["validTimeUtc", "ID_ESTACION"], right_on=["date", "ID_ESTACION"], how = "left")
```

```python
df_meteo_eto = df_meteo_eto.drop(["date"], axis=1)
```

```python
df_meteo_eto.to_csv("./ds_tractats/df_meteo_eto.csv", index = False)
```

```python
df_meteo_eto.columns
```

```python
preserve = ['validTimeUtc', 'ID_ESTACION']
preserve += [c for c in list(df_meteo_eto.columns) if "Avg" in c]
```

```python
new_ds = df_meteo_eto.loc[:,preserve]
```

```python
df_new = new_ds.copy().drop(columns = ['validTimeUtc', 'ID_ESTACION'])
df_columns = df_new.columns
```

```python
for c in df_columns:
    serie = (df_new.corrwith(df_new[c]) >0.98) & (df_new.corrwith(df_new[c]) != 1)
    if serie.sum() > 0:
        df_new = df_new.drop(columns=[c])
```

```python
df_new['validTimeUtc'] = df_meteo_eto['validTimeUtc'] 
df_new['ID_ESTACION'] = df_meteo_eto['ID_ESTACION']
```
